# Extracting bus coordinates

```python
import pandas as pd
import geopandas as gpd
```

```python
matrix = pd.read_csv('/distance_matrix_large', index_col = 0)
stops = gpd.read_file('/bus_stops', index_col = 0)
```

```python
ID = 6401544528 #where the number is 'osmid' from 'stops' gpd dataframe

#get the coordinates of stop by stop's OsmID
x = stops.loc[stops['osmid'] == ID].geometry.x.tolist()
y = stops.loc[stops['osmid'] == ID].geometry.y.tolist()

#coordinates of the bus stop
point = (y[0] , x[0])
```

### If visual check required

```python
import osmnx as ox
import networkx as nx
```

```python
stop = ox.graph_from_point(point,dist = 1000, 
    network_type='drive_service', simplify=False)
ox.plot_graph(stop, bgcolor = 'grey')
```
